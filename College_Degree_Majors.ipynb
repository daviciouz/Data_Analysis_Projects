** Introduction **: College Major Degrees from FivethirtyFive

```python
#Importing modules, data and getting it ready
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads = recent_grads.dropna()
```

** Using Scatter plots to explore relations in the data such as Sample_size and Median, Sample_size and Unemployment_rate, Full_time and Median, ShareWomen and Unemployment_rate, Men and Median, Women and Median **

With these plots, we want to be able to answer the following questions:
Do students in more popular majors make more money?
Do students that majored in subjects that were majority female make more money?
Is there any link between the number of full-time employees and median salary?

```python
#Sample_size and Median
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
#Sample_size and Unemployment_rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
#Full_time and Median
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

```python
#ShareWomen and Unemployment_rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
#Men and Median
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
#Women and Median
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

** Pandas, Histogram explore the distributions of the following columns:**
Sample_size
Median
Employed

** And to Answer below Questions **: 
What percent of majors are predominantly male? Predominantly female?
What's the most common median salary range?

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
	
```

```python
#experimenting with different bin sizes and ranges for the histograms.
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

** At this stage, we have an idea of certain relationships in our dataset, Therefore scatter matrix plot (scatter plots and histograms into one grid) would allow us to explore potential relationships and distributions simultaneously**

**Pandas, Scatter Matrix Plot**

Import scatter_matrix from pandas.tools.plotting in order to Create a 2 by 2 scatter matrix plot using the Sample_size and Median columns.

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6,6))
```

```python
#Now we create a 3 by 3 scatter matrix plot using the Sample_size, Median, and Unemployment_rate columns.
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
