**OVERVIEW**

In anticipation of "Star Wars: The Force Awakens to come out", The team at FiveThirtyEight conducted survey to guage Star Wars fans excitement. In particular, they wondered: does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?

```python
import pandas as pd
import numpy as np
```

```python
star_wars = pd.read_csv("StarWars.csv", encoding="ISO-8859-1")
```

Exploring dataset and removing any rows where RespondentID is NaN

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

Verifying the status of the cleaning done so far especially on the RespondentID column

```python
star_wars.head()
```

At this stage, we perform further cleaning and map the (YES/NO) Columns to Boolean Columns(TRUE/FALSE)

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

Normalizing the columns and renaming them to be more intuitive

```python


movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

Convert the next 6 columns to a numeric type and rename them to make our dataset more meaningful

```python

star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

 Calculating the mean of each of the ranking columns

```python
star_wars[star_wars.columns[9:15]].mean()
```

A bar chart of each ranking

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

**Ranking**

**SUMMARY I: After ranking the movie columns**


The computed the average ranking of each movie indicate that the "original" movies are rated much more highly than the newer ones.

Finding the Most viewed Movie

```python
star_wars[star_wars.columns[3:9]].sum()
```

**View Counts**

It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular

Splitting the Gender Column to derive different statistics by Gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

To derive Gender based statistics, we redo the two previous analyses (thus the viewed movie and the highest-ranked movie) separately for each group so we can compare the results.

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.show()

plt.bar(range(6), females[females.columns[9:15]].mean())
plt.show()
```

**Male/Female differences in favorite Star Wars movie and most seen movie**

Interestingly, more males watch episodes 1-3, but males liked them far less than females did.
